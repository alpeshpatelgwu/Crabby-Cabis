```python
import pytz

#Change to relative path
path = 'Weather txt' 
#Weather txt/724050-13743-2016

list_ = []

#Change to relative path!
# path = 'Weather txt' 

allFiles = glob.glob(path + "/*")
# df = pd.read_table(path + "/724050-13743-2016", delim_whitespace=True, names=('Year','Month','Day','Hour','Air Temp', 'Dew Point Temp', 'Sea Level Press.','Wind Dir.','Wind Speed Rate','Sky Cond.','1 hr Precip','6 Hour Precip'))
# df.head()

for file_ in allFiles:
    df = pd.read_table(file_, delim_whitespace=True, \
                       names=('Year','Month','Day','Hour','temp',\
                              'precip'),\
                      parse_dates={"ts": [0, 1, 2, 3]},\
                      na_values="-9999",\
                      infer_datetime_format=True,\
                      usecols=[0, 1, 2, 3, 4, 10])
    list_.append(df)

frame = pd.concat(list_, axis = 0, ignore_index = True)


def correct_tz(x):
    x = x.replace(tzinfo=pytz.utc).astimezone(
        pytz.timezone("US/Eastern")).replace(tzinfo=None)
    return x

frame["ts"] = frame["ts"].apply(correct_tz)

# Change values of -1 in precip (meaning "trace percipitation") to
# 0.1.
frame["precip"] = frame["precip"].apply(
    lambda x: 0.1 if x == -1 else x)

# Remove x10 multiplier from temperature and precipitation values.
frame["temp"] = frame["temp"].apply(lambda x: x / 10)
frame["precip"] = frame["precip"].apply(lambda x: x / 10)
        
frame.head()
```

```python
frame.to_sql(name='hourly_weather', if_exists='append', con=conn, chunksize=500000, index=False)
```
