```python
#Import dependencies

import pandas as pd
import json
from shapely.geometry import shape, Point
# depending on your version, use: from shapely.geometry import shape, Point


```

```python
#Import CSV

locations_df = pd.read_csv('Capital_Bike_Share_Locations.csv')
locations_df["NUMBER_OF_BIKES"] = locations_df["NUMBER_OF_BIKES"] + locations_df["NUMBER_OF_EMPTY_DOCKS"]
locations_df = locations_df.drop(columns=["NUMBER_OF_EMPTY_DOCKS", "INSTALLED", "LOCKED", "INSTALL_DATE", "REMOVAL_DATE", "TEMPORARY_INSTALL", "SE_ANNO_CAD_DATA", "X", "Y"])
locations_df["Ward"] = ""
locations_df
```

```python
# load GeoJSON file containing sectors
with open('../dc-ward-map-overlay.geojson') as f:
    js = json.load(f)

# construct point based on lon/lat returned by geocoder
#point = Point(-122.7924463, 45.4519896)

# check each polygon to see if it contains the point

for index, row in locations_df.iterrows():
    point = Point(row['LONGITUDE'],row['LATITUDE'])
    #print(point)
    for feature in js['features']:
        polygon = shape(feature['geometry'])
        #print(polygon)
        #if point.within(polygon):
        if polygon.contains(point):
            print(point)
            print(feature["properties"]["name"])
            #row["Ward"] = feature["properties"]["name"]
            locations_df.loc[index,'Ward'] = feature["properties"]["name"]
```
